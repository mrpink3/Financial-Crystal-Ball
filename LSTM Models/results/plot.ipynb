```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
test_results = pd.read_csv("test_predictions_Regression.csv")
```

```python
test_results
```

```python
# Plot the predictions vs actual returns
plt.figure(figsize=(14, 7))
plt.plot(test_results['True'], label='Actual Return', color='blue', alpha=0.5)
plt.plot(test_results['Predicted'], label='Predicted Return', color='red', alpha=0.5)
plt.title('LSTM Regression: Actual vs Predicted Returns')
plt.xlabel('Sample Index')
plt.ylabel('Return')
plt.legend()
plt.grid()
plt.show()
```
