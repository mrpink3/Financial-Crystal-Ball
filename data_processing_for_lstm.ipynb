```python
import pandas as pd

sentiment_data = pd.read_csv("sentiment analysis/daily_finbert_aggregated_with_class.csv")
sentiment_data[sentiment_data['date']>='2004-08-19'].head(10)
```

```python
finance_data = pd.read_csv("data/final_dataset.csv")
finance_data.head(10)
```

```python
# Rename  the 'date' column in sentiment_data to match finance_data
sentiment_data.rename(columns={'date': 'Date'}, inplace=True)

#Merge the two dataframes on the 'date' column
merged_data = pd.merge(finance_data, sentiment_data, on='Date', how='inner')
```

```python
merged_data.head(1)
```

```python
# Create a new dataframe called Disney_data
# It consists of 12 columns the 'Date', 'Day 1-10', 'sentiment' columns
# 'Day 1-10' columns are the 10 days before the 'Date' column

Disney_data = pd.DataFrame()
Disney_data['Date'] = merged_data['Date']
# Create 'Day 1-10' columns
for i in range(1, 11):
    Disney_data[f'Day {i}'] = merged_data['Disney'].shift(i)
# Create 'sentiment' column
Disney_data['sentiment'] = merged_data['mean_score']
```

```python
# Drop rows with NaN values
Disney_data.dropna(inplace=True)
# Reset the index
Disney_data.reset_index(drop=True, inplace=True)

# Rename mean_score column to sentiment
Disney_data.rename(columns={'mean_score': 'sentiment'}, inplace=True)
```

```python
df = Disney_data.copy()

price_columns = [f'Day {i}' for i in range(1, 11)]

# Compute return: (next_day_price - current_price) / current_price
df['today_price'] = df['Day 10']
df['next_price'] = df['Day 10'].shift(-1)
df['next_return'] = (df['next_price'] - df['today_price']) / df['today_price']

# Drop rows where next_return is NaN (last row)
df = df[:-1]

# Create movement label (classification target)
def get_movement_label(r):
    if r > 0.005:
        return 2
    elif r < -0.005:
        return 0
    else:
        return 1

df['movement_label'] = df['next_return'].apply(get_movement_label)

# Final columns: Day1-Day10, Sentiment, movement_label / next_return
classification_df = df[price_columns + ['sentiment', 'movement_label']]
regression_df = df[price_columns + ['sentiment', 'next_return']]

# Save to CSV
classification_df.to_csv("LSTM Models/data/classification_data.csv", index=False)
regression_df.to_csv("LSTM Models/data/regression_data.csv", index=False)
```
