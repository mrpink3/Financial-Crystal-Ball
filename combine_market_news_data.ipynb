```python
import pandas as pd

# Dosya yolları
market_path = "data/combined/market_data_interpolated.csv"
news_path = "data/news/processed/nyt_daily_filtered.csv"

# Veri setlerini oku
market_df = pd.read_csv(market_path, parse_dates=["Date"], index_col="Date")
news_df = pd.read_csv(news_path, parse_dates=["date"], index_col="date")

news_df.index.name = "Date"  # İki index de artık "Date"
```

```python
# Tarih bazında birleştirme
combined_df = market_df.join(news_df, how="left")

# Sonuç
combined_df.head()
```

```python
combined_df.to_csv("data/final_dataset.csv")
print("✅ Birleştirme tamamlandı ve kaydedildi.")
```
